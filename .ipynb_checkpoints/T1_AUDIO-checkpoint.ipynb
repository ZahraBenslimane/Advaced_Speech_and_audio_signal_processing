```python
antenne=array('play')     # When playing recorded files (can work without acquisition system)
```

```python
# Load acquisition and array parameters from the antenne variable, AFTER launching acquisition or play
Fs = antenne.fs
BLK = antenne.blocksize
N = antenne.mems_nb
d = antenne.interspace
```

```python
# Read an audio buffer
m = antenne.read()
```

```python
print(m)
print(m.shape)
```

```python
#400Hz
m0 = m[0,:]
m7 = m[7,:]

t = (1/Fs)*np.arange(len(m0))

plt.figure()
plt.grid(visible=True, which='both', color='0.65', linestyle='-')
plt.plot(t,m0, 'g', label ='M0')
plt.plot(t,m7, 'r', label = 'M7')

t1 = 0.00143
t2 = 0.00267 

plt.plot(t2,0.24, 'or')
plt.plot(t1,2.69, 'og')

plt.title('Display the waveforms recorded by the first and the last microphones ')
plt.xlim([0 , 0.011])
plt.minorticks_on()
plt.legend()
```

```python
# Calcul de TFD du signal 
M0 = np.fft.fft(m0)

plt.figure(figsize = (10,8))
plt.subplot(2,2,1)
f = np.linspace(0, Fs,len(m0))
plt.plot(f, np.abs(M0)/len(m0))
plt.title("Amplitude du spectre du signal M0" , fontsize = 8)
plt.xlabel("Frequency(Hz)", fontsize = 7)
plt.ylabel("$|TFD(m0)|$")
plt.xlim(0, 1000)
plt.grid(True)  

plt.subplot(2,2,2)
plt.phase_spectrum(m0, Fs=Fs, color='C1')
plt.xlabel("Frequency(Hz)" , fontsize = 7)
plt.title("Phase du Spectre du signal M0", fontsize = 8)
plt.grid(True)

# Micro 7
M7 = np.fft.fft(m7)

plt.subplot(2,2,3)
f = np.linspace(0, Fs,len(m7))
plt.plot(f, np.abs(M7)/len(m7))
plt.title("Amplitude du spectre du signal M7", fontsize = 8)
plt.xlabel("Frequency(Hz)" , fontsize = 7)
plt.ylabel("$|TFD(m7)|$")
plt.xlim(0, 1000)
plt.grid(True)  

plt.subplot(2,2,4)
plt.phase_spectrum(m7, Fs=Fs, color='C1')
plt.xlabel("Frequency(Hz)" , fontsize = 7)
plt.title("Phase du Spectre du signal M7", fontsize = 8)
plt.grid(True)
```

### 7. Repeat the previous steps with a white noise sound source. Calculate the Discrete Fourier Transforms (DFT) (modulus + phase) of the waveforms recorded by the first and the last microphones in the array. Propose a method using the phase difference of the two DFTs to estimate the speed of sound.

```python
antenne_noise=array('play')     # When playing recorded files (can work without acquisition system)
```

```python
# Load acquisition and array parameters from the antenne variable, AFTER launching acquisition or play
Fs = antenne_noise.fs
BLK = antenne_noise.blocksize
N = antenne_noise.mems_nb
d = antenne_noise.interspace
```

```python
# Read an audio buffer
noise = antenne_noise.read()
```

```python
# Calcul de TFD du signal 

noise_m0 = noise[0,:]
noise_m7 = noise[7,:]

FFT_noise_m0 = np.fft.fft(noise_m0)

plt.figure(figsize = (10,8))
plt.subplot(2,2,1)
f = np.linspace(0, Fs,len(noise_m0))
plt.plot(f, np.abs(FFT_noise_m0)/len(noise_m0))
plt.title("Amplitude du spectre du signal FFT_noise_m0" , fontsize = 8)
plt.xlabel("Frequency(Hz)", fontsize = 7)
plt.ylabel("$|TFD(noise_m0)|$")
plt.xlim(0, 5000)
plt.grid(True)  

plt.subplot(2,2,2)
plt.phase_spectrum(noise_m0, Fs=Fs, color='C1')
plt.xlabel("Frequency(Hz)" , fontsize = 7)
plt.title("Phase du Spectre du signal FFT_noise_m0", fontsize = 8)
plt.grid(True)

# Micro 7
FFT_noise_m7 = np.fft.fft(noise_m7)

plt.subplot(2,2,3)
f = np.linspace(0, Fs,len(noise_m7))
plt.plot(f, np.abs(FFT_noise_m7)/len(noise_m7))
plt.title("Amplitude du spectre du signal FFT_noise_m7", fontsize = 8)
plt.xlabel("Frequency(Hz)" , fontsize = 7)
plt.ylabel("$|TFD(noise_m7)|$")
plt.xlim(0, 5000)
plt.grid(True)  

plt.subplot(2,2,4)
plt.phase_spectrum(noise_m7, Fs=Fs, color='C1')
plt.xlabel("Frequency(Hz)" , fontsize = 7)
plt.title("Phase du Spectre du signal FFT_noise_m7", fontsize = 8)
plt.grid(True)
```

```python
def compute_rms(chunk):
	try:
		len(chunk)
		chunk = pow(abs(chunk),2)
		return math.sqrt (chunk.mean())
	except TypeError: return chunk
    
    
RMS_m400Hz_m7 = compute_rms(m400Hz_m7)    
print("RMS_m400Hz_m7 = ", RMS_m400Hz_m7)
```

#  Generate a pure tone at a frequency below the limit frequency previously measured. Record and display the waveforms measured by each microphone of the array

```python
# All required import
import matplotlib.pyplot as plt
import ipywidgets as widgets
import numpy as np
from client import array
import time
import math
%matplotlib notebook

antenne_400Hz_face_centre=array('play')     # When playing recorded files (can work without acquisition system)
```

```python
# Load acquisition and array parameters from the antenne variable, AFTER launching acquisition or play
Fs = antenne_400Hz_face_centre.fs
BLK = antenne_400Hz_face_centre.blocksize
N = antenne_400Hz_face_centre.mems_nb
d = antenne_400Hz_face_centre.interspace

# Read an audio buffer
m400Hz_centre46cm = antenne_400Hz_face_centre.read()
```

```python
plt.figure()
tR =  (1/Fs)*np.arange(len(m400Hz_centre46cm[0]))
plt.grid(visible=True, which='both', color='0.65', linestyle='-')
plt.plot(tR,m400Hz_centre46cm[0],  label = "M0") 
plt.plot(tR,m400Hz_centre46cm[1],  label = "M1") 
plt.plot(tR,m400Hz_centre46cm[2],  label = "M2") 
plt.plot(tR,m400Hz_centre46cm[3],  label = "M3")
plt.plot(tR,m400Hz_centre46cm[4],  label = "M4") 
plt.plot(tR,m400Hz_centre46cm[5],  label = "M5") 
plt.plot(tR,m400Hz_centre46cm[6],  label = "M6") 
plt.plot(tR,m400Hz_centre46cm[7], 'b', label = "M7") 

plt.title('Waveforms recorded 46cm from centre of microphone array at 400Hz : ')
plt.minorticks_on()
plt.legend()
#plt.xlim([0.0521,0.0498])
```

```python
# On choisi deux instants:
K1 = int(0.0521*Fs)
K2 = int(0.0498*Fs)

# On extrait les signaux dans ces deux instants
chucks = []
for i in range(8):
    chucks.append(list(m400Hz_centre46cm[i][K1:K2]))
        
T = np.argmax(chucks, axis = 1)/Fs  
print("Le time of the max in the chosen chunk :", T)
```

```python
T0 = T[4]
delays = T - T0
for i in range(8):
    print("Time Delay between microphone ",i+1,"and 5 : ",delays[i], "s")
```

### Interpret the shape of the curve ⌧n (defined in the preparation) as a function of zn, coordinates of the microphones. Conclude on the approximation to adopt on the shape of the wavefront.

```python
d = 0.06 
N = 8
n = np.arange(1,9,1)
theta = np.pi/6
zn = d*(n-N/2) - d/2
Tau_n = -zn * np.cos(theta)/340

plt.figure(figsize = (8,5))
plt.plot(zn, Tau_n)
plt.title("The delay tau_n needed by the wave to travel the distance d_n\n as a function of the coordinates of the microphones")
plt.xlabel("zn : coordinates of the microphones")
plt.ylabel("tau_n")
```

```python
plt.figure(figsize = (8,5))
zn = d*(n-N/2) - d/2
plt.plot(zn, delays)
plt.title("Experimental delay of sound travel as a function of the position of the microphone")
plt.xlabel("zn : coordinates of the microphones")
plt.ylabel("Tau_n")
```
