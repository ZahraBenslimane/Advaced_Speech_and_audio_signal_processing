```python
#  (a) the observation of the waveform :
# à partir des tracés des waveforme on extrait deux pics successifs dans les deux signaux et on fait la différence temporel 
# pour calculer le retard
t1 = 0.08651
t2 = 0.08762

To = t2 - t1
print("le retard à partir des observation est : To = ",1000*To, 'ms')
```

```python
#  (b) the phase of the spectra
# on a la relation entre le retard est le déphasage
# Dphi = 2*pi*f0
```

 Formule pour le calcul du retard à partir du déphasage
 $\Delta t = \frac{\Delta \phi}{2 \pi f_0}$
 
 $ f_0 $ est la fréquence du signal audio émis

```python
# extraction de la phase à la fréquence cible dans le sign
#Fl = 400
#kl = round(Fl*len(XRf1)/Fs)
#pha1 = np.angle(XRf1[kl]) # il y a un probleme psq il fait un modulo 2pi

# extraction de la phase à la fréquence cible dans le sign
#pha2 = np.angle(XRf[kl])  # il y a un probleme psq il fait un modulo 2pi
#print(pha2)

# Solution 
# extraction des phases à partir du tracé du spectre : !! la spectre dessiné est en radians

ph1 = -2.4 # rad  mic1
ph2 = -5.39 # rad  mic8

dephasage =  np.abs(ph2-ph1) 
print(dephasage)
# frequence du son émi 400 Hz
dT = dephasage/(2*np.pi*400)
print("valeur du retard à partir du déphasage : retard = ",dT*1000, "ms")
```

```python
# 6. Deduce the speed of sound propagation, also known as the celerity. Compare with the
# theoretical value at room temperature.

d = 41*10**(-2) # longueur de l'antenne en mettre
c = d/To
print("vitesse obtenu à partir des observation des waveforme,  celerity = ",c, 'm/s')
```

```python
# calcul de la vitesse du son à partir du retard obtenu par déphasage
c1 = d/dT
print("vitesse obtenu à partir du déphasage, celerity = ",c1, 'm/s')
```

```python
# comparaison avec la vitesse du son dans une salle à température ambiante :
c0 = 340 # m/s

#erreur de mesure
err = abs(c-c0)/c0
err1 = abs(c1-c0)/c0

print("erreur de mesure à partir de l'observation des waveforme : err = ",err*100, " %")
print("erreur de mesure à partir du déphasage : err = ",err1*100, " %")
```

### Tracé de la courbe théorique du retard en fonction de zn

```python

tet_s = np.pi*0.45
d = 0.06 #m
N = 8

n = np.arange(1,9,1)
zn = d*(n-N/2) - d/2
print(zn)
Tn = -zn * np.cos(tet_s)/340
print(Tn)

```
